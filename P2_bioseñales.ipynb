1. Implemente una función en Python que permita generar una función escalón unitario, definida en un intervalo $n_1≤n_0≤n_2$

```python
def escalon_unitario(n0,n1,n2):
    n = np.arange(n1, n2 + 1)
    x = ((n - n0) >= 0).astype(int)
    return x,n
```

```python
x,n=escalon_unitario(0,-5,5)
plt.figure(figsize=(20,3))
plt.stem(n,x)
plt.xticks(np.arange(np.min(n),np.max(n)+1))
plt.xlabel("Muestra")
plt.ylabel("Amplitud")
plt.grid()
plt.show()
```

2. Implemente una función en Python que permita generar una función rampa, definida en un intervalo $n_1≤n_0≤n_2$ con una pendiente m.

```python
def rampa(n0,n1,n2,m):
    n=np.arange(n1,n2+1) #Vector de muestras
    u=(n-n0)>=0
    w=(n-n0)*u
    wp=w*m #Pendiente
    return wp,n,m
```

```python
r,nr,m=rampa(0,-5,5,1)
plt.figure(figsize=(20,3))
plt.plot(nr,r)
plt.xlabel("Muestra")
plt.ylabel("Amplitud")
plt.xticks(np.arange(np.min(nr),np.max(nr)+1))
plt.grid()
plt.show()
```

3. Genere las siguientes secuencias usando las funciones básicas de Python que se han presentado. Grafique los resultados.

	a. $x_1 (n)= 3δ(n+1)+5δ(n+3)+3δ(n+2)+3δ(n)+ δ(n)$. ¿Cómo debe ser el vector de muestras?


```python
def impulso(n0,n1,n2,amp):
    n=np.arange(n1,n2+1)
    x=(n-n0)==0
    return amp*(x.astype(int)),n
```

```python
n1=-5
n2=5
eq1x,eqn=impulso(-1,n1,n2,3)
eq2x,_=impulso(-3,n1,n2,5)
eq3x,_=impulso(-2,n1,n2,3)
eq4x,_=impulso(0,n1,n2,3)
eq5x,_=impulso(0,n1,n2,1)
x1=eq1x+eq2x+eq3x+eq4x+eq5x
print(x1)
plt.figure(figsize=(20,3))
plt.stem(eqn,x1)
plt.xlabel("Muestra")
plt.ylabel("Amplitud")
plt.title(r"$x_1 (n)$= 3δ(n+1)+5δ(n+3)+3δ(n+2)+3δ(n)+ δ(n)")
plt.grid()
plt.xticks(np.arange(np.min(eqn),np.max(eqn)+1))
plt.show()
```

c. El siguiente sistema arroja una secuencia definida entre  0≤n≤100

```python
def seno(freq,theta):
    n=np.arange(0,101)
    w=2*np.pi*freq
    x=np.sin((w*n)+theta)
    return(x)
def expo(sigma,jw0):
    n=np.arange(0,101)
    x=np.exp((sigma+jw0)*n)
    return(x)


eq1=seno(1,0)
eq2=expo(0.01,0)
x3=eq1*eq2
print(x3)
plt.figure(figsize=(20,3))
plt.title(r"$x_3(n)$=sin($2 \pi n$) x $e^{0.01n}$")
plt.plot(np.arange(0,101),x3)
plt.xlabel("Muestra")
plt.ylabel("Amplitud")
plt.grid()
plt.show()
```

```python
x1,nx1,_=rampa(0,0,5,1)
x2,nx2,_=rampa(5,6,11,1)
x3,nx3,_=rampa(10,12,17,1)
rango=np.arange(0,17)
der1=np.diff(x1)
der2=np.diff(x2)
der3=np.diff(x3)
plt.figure(figsize=(20,10))
plt.subplot(2,1,1)
plt.stem(nx1,x1,linefmt='black')
plt.stem(nx2,x2,linefmt='blue')
plt.stem(nx3,x3,linefmt='red')
plt.title('Rampas Originales')
plt.grid()
plt.xlabel('Muestra')
plt.ylabel('r[n]')
plt.subplot(2,1,2)
plt.stem(nx1[1:],der1,linefmt='black')
plt.stem(nx2[1:],der2,linefmt='blue')
plt.stem(nx3[1:],der3,linefmt='red')
plt.title('Derivadas de las Rampas')
plt.xlabel('Muestra')
plt.ylabel('Derivada')
plt.grid()
plt.show()
```

**Observación:** La derivada discreta es una diferencia finita y no es exacta como en el caso continuo, muestra saltos abruptos donde hay cambios de tramo, la derivada discreta no captura exactamente la tasa de cambio instantánea como lo hace la derivada continua.
