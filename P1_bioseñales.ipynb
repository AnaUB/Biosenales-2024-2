1. Escribir una función que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota mínima, la máxima, media y la desviación típica.


```python
notas = {'José': 9, 'María': 6.5, 'Miguel': 8.2, 'Ana': 7.8}
```

```python
def calcular_estadisticas(dic):
    serie = pd.Series(dic)
    return pd.Series({
        'Mínima': serie.min(),
        'Máxima': serie.max(),
        'Media': serie.mean(),
        'Desviación estándar': serie.std()
    })
```

```python
calcular_estadisticas(notas)
```

2. Cálculo del Índice de Masa Corporal (IMC) con Pandas.

**Instrucciones**

a. Cargar los datos desde el archivo "datos.csv".

b. Mostrar las primeras y últimas filas del DataFrame para conocer la estructura de los datos.

c. Borrar la columna "Unnamed: 0" del DataFrame.

d. Calcular el IMC con la fórmula:

    IMC= Peso (kg)/(Altura (m)^2)

Nota: La altura está en centímetros, por lo que se debe dividir entre 100 para obtener metros.

e. Agregar el IMC como una nueva columna llamada "BMI" en el DataFrame.

f. (Opcional): Categorizar a los participantes según su IMC en:
- Bajo peso (< 18.5)
- Normal (18.5 - 24.9)
- Sobrepeso (25 - 29.9)
- Obesidad (≥ 30)


```python
df = pd.read_csv("datos.csv")
# Se intercambiaron las columnas "Height" y "Weight" en el datatset original para que los datos tuvieran más sentido
```

```python
print("Primeras filas: ")
display(df.head())

print("\n Últimas filas: ")
display(df.tail())
```

```python
if 'Unnamed: 0' in df.columns: # Para evitar errores en caso de queya se haya eliminado la columna
    df.drop(columns=['Unnamed: 0'], inplace=True)
df
```

```python
df["Height_m"] = df["Height"] / 100
df["BMI"] = df["Weight"] / (df["Height_m"] ** 2)
df["BMI"]
```

```python
def categorizar_imc(imc):
    if imc < 18.5:
        return 'Bajo peso'
    elif 18.5 <= imc < 25:
        return 'Normal'
    elif 25 <= imc < 30:
        return 'Sobrepeso'
    else:
        return 'Obesidad'

df['Categoría_IMC'] = df['BMI'].apply(categorizar_imc)
df['Categoría_IMC'].value_counts()
```
